```python
lst = [['Nam',18,'Q2'],['An',19,'Q5'],['Hoa',20,np.nan]]
df1 = pd.DataFrame(lst,columns=['Name','Age','Address'])
df1
```

```python
lst = [['Nam','DS1'],['An','DS2'],['Trung','DS2']]
df2 = pd.DataFrame(lst,columns=['Ten','Class'])
df2
```

```python
df = pd.concat([df1,df2],join='outer',axis=1)
df
```

```python
df = pd.concat([df1,df2],join='inner',axis=1)
df
```

```python
df = pd.merge(df1,df2,how='outer')
df
```

```python
df = pd.merge(df1,df2,left_on='Name',right_on='Ten',how='inner')
df
```

```python
df = pd.merge(df1,df2,how='left')
df
```

```python
df = pd.merge(df1,df2,how='right')
df
```

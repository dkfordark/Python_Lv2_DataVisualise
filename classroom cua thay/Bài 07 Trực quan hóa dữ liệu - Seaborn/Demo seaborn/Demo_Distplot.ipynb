```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('schoolimprovement2010grants.csv')
data.head()
```

```python
data['region'] = data['Location'].map(lambda x : "South" if "ST" in x else "west")
```

```python
data.head()
```

```python
data.info()
```

```python
#create displot
plt.figure(figsize=(10,6))
sb.distplot(data['Award Amount'],bins=20,color='purple',kde=False)
#display a plot
plt.show()
```

```python
#create displot
plt.figure(figsize=(10,6))
sb.distplot(data['Award Amount'],hist=False,rug=True)
#display a plot
plt.show()
```

```python
#create displot
plt.figure(figsize=(10,6))
sb.distplot(data['Award Amount'],hist=False,rug=True, kde_kws={'shade':'kde_kws'})
#display a plot
plt.show()
```

```python
data['region'].value_counts()
```

```python
df_west = data[data['region'] == "west"]
df_south = data[data['region'] == "South"]
```

```python
plt.hist(df_west['Award Amount'], label='west',alpha=0.5)
plt.hist(df_south['Award Amount'], label='south',alpha=0.5)
plt.legend()
plt.xlabel('Award Amount')
plt.show()
```

```python
#create displot
plt.figure(figsize=(10,6))
sb.distplot(df_west['Award Amount'],hist=False,label='west')
sb.distplot(df_south['Award Amount'],hist=False,label='south')
#display a plot
plt.show()
```

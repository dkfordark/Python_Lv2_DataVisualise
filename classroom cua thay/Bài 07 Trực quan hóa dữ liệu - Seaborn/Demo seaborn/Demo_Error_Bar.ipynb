## Phân tích dữ liệu Categorical

```python
# How many male-female in ours sample
sb.countplot(x='sex',data=df)
```

```python
# How about total_bill vs Male-Female
sb.swarmplot(x='sex',y='total_bill',data=df)
```

```python
# How about tips vs Male-Female
sb.swarmplot(x='sex',y='tip',data=df)
```

```python
# How about Male-Female vs smoker
sb.countplot(x='sex',data=df,hue='smoker')
```

```python
sb.boxplot(x='day',y='total_bill',data=df)
```

## Demo palette color

```python
for st in ['white','dark','whitegrid','darkgrid','ticks']:
    sb.set_style(st)
    sb.barplot(x='day',y='total_bill',data=df)
    plt.show()
```

```python
# lấy palette màu hiện hành
#colpal = sb.color_palette()
#sb.palplot(colpal)
#sb.set()
sb.set_palette('pastel')
sb.barplot(x='day',y='total_bill',data=df)
```

```python
# Các color palette :
# Categorical palette: deep, muted, pastel, bright, dark, colorblind
for b in sb.palettes.SEABORN_PALETTES:
    sb.palplot(sb.color_palette(b))
    plt.show()
```

```python
# màu phân kỳ coolwarm, RdBu_r,BrBG
colpal = sb.color_palette(palette='coolwarm', n_colors=12)
#hiển thị bảng màu
#sb.palplot(colpal)
sb.set_palette(colpal)
sb.barplot(x='day',y='total_bill',data=df)
```

```python
# màu tuần tự
colpal = sb.color_palette(palette='Blues_r',n_colors=10)
sb.palplot(colpal)
```

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#prepare data
year = [1950,1951,1952,1953,1954]
pop = [2.53, 2.57, 2.62, 2.67, 2.71]
```

```python
#using pyplot

#plot data
plt.plot(year,pop, label="Population")

#add legend - title
plt.title('Population from 1950 -1954', fontsize=14)
plt.legend(title='Legend')


#set parameters
plt.xticks(year)
plt.xlabel('Year')
plt.ylim(2.5,2.75)

# show the plot
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

#plot data 
ax.plot(year,pop, label="Population")

#add legend - title
ax.set_title('Population from 1950 -1954', fontsize=14)
ax.legend(title='Legend')

#set parameters
ax.set_xticks(year)
ax.set_xlabel('Year')
ax.set_ylim(2.5,2.75)

# show the plot
plt.show()
```

```python
#plot using blue circle markers

plt.plot(year,pop,'rD-.')
plt.xticks(year)
plt.show()
```

```python
fig = plt.figure(figsize=(14,8))
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x - 0), color='blue', label='sin(x)')   #specify color by name
plt.plot(x,np.sin(x - 1), color='r', label='sin(x-1)')   #specify color by name  
plt.plot(x,np.sin(x - 2), color='#FFDD44', label='sin(x-2)')   #specify color by name  RGB
plt.plot(x,np.sin(x - 3), color=(0.2,0.3,0.5), label='sin(x-3)')   #specify color by name 
plt.plot(x,np.sin(x - 4), color='chartreuse', label='sin(x-4)')   #specify color by name


plt.legend(title='Ghi chu')

plt.show()
```

```python
#line with red diamond
plt.plot(pop,'rd--')
plt.show()
```

```python
# more line format
plt.plot(year,pop, color='green', marker='o',linestyle='dashed',linewidth=2, markersize=12)
plt.title('Biểu đồ dân số', color='red')
plt.ylabel('Dân số',color='red')
plt.xlabel('Năm',color='red')
plt.xticks(year)
plt.show()
```

```python
#plotting multiple sets of data
plt.plot([1,2,3], label='line 1', marker='*', markersize=10)
plt.plot([1.5,2.5,3.5], label='line 2', marker='o', markersize=10)
plt.title('Biểu đồ song song', color='red', fontsize=15)
plt.ylabel('y value',color='red')
plt.legend(title='Ghi chu')
plt.grid(True)
plt.show()
```

```python
#demo area
x = range(1,15)
y = [1, 4, 6, 8, 4,5,3,2,4,1,5,6,8,7]
plt.fill_between(x,y)
plt.show()
```

```python
#demo stack area chart
x = range(1,15)
y = np.array([1, 4, 6, 8, 4,5,3,2,4,1,5,6,8,7])
y1 = y +3
plt.stackplot(x,y1,y,alpha=0.5)
plt.show()
```

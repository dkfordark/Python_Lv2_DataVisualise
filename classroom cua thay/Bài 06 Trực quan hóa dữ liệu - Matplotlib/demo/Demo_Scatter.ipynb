# Histogram and Normal Distribution

```python
mu, sigma = 0,1
s = np.random.normal(mu,sigma,1000)

#create bins and histogram
count, bins, ignored = plt.hist(s,35,density=True)

#plot the distribution curve
plt.plot(bins, 1/(sigma * np.sqrt(2*np.pi)) * np.exp(- (bins - mu)**2 / (2 * sigma**2)),linewidth=3,color='black')
plt.show()
```

```python
# Import thư viện
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [26,37,35,53]
labels = ['10A1','10A2','10A3','10A4']
```

```python
plt.pie(x,labels= labels)
plt.show()
```

```python
plt.pie(x,labels= labels,counterclock=False,shadow = True, autopct='%.0f%%',startangle=90)
plt.title("Số học viên trong lớp")
plt.legend(loc=1)
plt.show()
```

```python
#ie chart
labels = ['Frogs','Hogs','Dogs','Logs']
sizes = [15,30,45,10]
explode = (0,0.2,0,0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)
ax1.axis('equal')
plt.show()
```
